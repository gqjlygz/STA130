In a simple linear regression model, we aim to model the relationship between a predictor (or independent variable)X, and an outcome (or dependent variable)Y

- Predictor Variable: This is an independent variable, chosen arbitrarily or observed, that we believe might influence the outcome.

- Outcome Variable: This is the dependent variable, which we are interested in modeling or predicting.

- Intercept: This is the value of the outcome when the predictor is zero. It is the "starting point" of the line on the vertical axis.

- Slope: This represents the rate of change in the outcome relative to changes in the predictor. It determines the steepness of the line.

- Error Term: This term represents random noise or variability in the outcome that isn't explained by the predictor. Itâ€™s generally assumed to follow a normal distribution with a mean of 0 and a certain standard deviation.

The following code demonstrates a simple linear regression model by generating a predictor variable and calculating the outcome based on chosen values for intercept, slope, and error term standard deviation. The error term is generated from a normal distribution.

```python
import numpy as np
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt

# Set parameters for the model
n = 100  # Number of samples
intercept = 2  # Intercept
slope = 3  # Slope
error_std_dev = 1.5  # Standard deviation of error term

# Generate predictor variable (uniformly distributed)
predictor = uniform.rvs(0, 10, size=n)

# Generate error term (normally distributed)
error = norm.rvs(0, error_std_dev, size=n)

# Generate outcome variable using the linear model equation
outcome = intercept + slope * predictor + error

# Plot the data and the theoretical regression line
plt.scatter(predictor, outcome, label='Observed data')
plt.plot(predictor, intercept + slope * predictor, color='red', label='True regression line')
plt.xlabel('Predictor')
plt.ylabel('Outcome')
plt.legend()
plt.title('Simple Linear Regression Model')
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
from scipy.stats import norm

# Load Old Faithful data, restrict to long wait times
import seaborn as sns
old_faithful = sns.load_dataset('geyser')
long_wait = old_faithful['waiting'] >= 62
long_wait_data = old_faithful[long_wait]

# Number of bootstrap samples
n_bootstrap = 1000
bootstrapped_slopes = []

# Bootstrap sampling process
for _ in range(n_bootstrap):
    # Resample data with replacement
    bootstrap_sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    
    # Fit the Simple Linear Regression model to the bootstrap sample
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    
    # Collect the slope coefficient from the fitted model
    bootstrapped_slopes.append(model.params['waiting'])

# Convert to DataFrame for plotting
bootstrap_df = pd.DataFrame(bootstrapped_slopes, columns=['slope'])

# Plot the sampling distribution of the slope coefficient
fig = px.histogram(bootstrap_df, x='slope', nbins=30,
                   title="Bootstrapped Sampling Distribution of the Slope Coefficient",
                   labels={'slope': 'Slope Coefficient'})
fig.show(renderer="png")

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrapped_slopes, 2.5)
upper_bound = np.percentile(bootstrapped_slopes, 97.5)
print(f"95% Confidence Interval for Slope: [{lower_bound:.4f}, {upper_bound:.4f}]")

# Check if 0 is within the confidence interval
contains_zero = lower_bound <= 0 <= upper_bound
print(f"Does the confidence interval contain 0? {'Yes' if contains_zero else 'No'}")

# Calculate p-value for the original data using the fitted model
original_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
original_slope = original_model.params['waiting']
print("Original model slope:", original_slope)
print(original_model.summary().tables[1])
```

**(A) Permutation Test**

- A permutation test allows us to test the null hypothesiswhich states that there is no difference in the average eruption duration between "short" and "long" wait groups. We do this by randomly shuffling the "kind" labels (i.e., "short" and "long") and recalculating the mean difference in durations multiple times. This generates a distribution of mean differences under the null hypothesis. We then compare the observed mean difference to this distribution to determine if it is statistically significant.

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')
old_faithful['kind'] = old_faithful['waiting'].apply(lambda x: 'short' if x < 68 else 'long')

# Observed mean difference
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test
n_permutations = 1000
permuted_diffs = []
for _ in range(n_permutations):
    shuffled = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1).values)
    permuted_diff = shuffled.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(permuted_diff)

# Calculate p-value
p_value = (np.sum(np.abs(permuted_diffs) >= np.abs(observed_diff)) + 1) / (n_permutations + 1)
print("Observed Mean Difference:", observed_diff)
print("P-Value from Permutation Test:", p_value)
```

**(B) Create a 95% bootstrap confidence interval by repeatedly bootstrapping**

- A bootstrap confidence interval estimates the range within which the true difference in means likely falls by resampling from each group with replacement. For each bootstrap sample, we calculate the mean difference between "short" and "long" wait times and then find the 2.5th and 97.5th percentiles of these differences to form a 95% confidence interval.

```python
# Number of bootstrap samples
n_bootstrap = 1000
bootstrapped_diffs = []

for _ in range(n_bootstrap):
    resampled = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    bootstrapped_diff = resampled.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_diffs.append(bootstrapped_diff)

# Calculate 95% confidence interval
ci_lower, ci_upper = np.percentile(bootstrapped_diffs, [2.5, 97.5])
print("95% Bootstrap Confidence Interval for Mean Difference:", (ci_lower, ci_upper))
```

**Explanation of Sampling Approaches**

- **Permutation Test:** The permutation test works by randomly shuffling the labels (i.e., "short" and "long") for each observation, simulating the null hypothesis of no difference. By repeating this shuffle many times, we create a distribution of mean differences under the null hypothesis. The observed mean difference is then compared to this distribution to assess whether it is significantly different.

- **Bootstrap Confidence Interval:** Bootstrapping involves repeatedly resampling with replacement within each group (i.e., "short" and "long"). For each resample, we calculate the mean difference between groups, and after many resamples, we use the distribution of these differences to calculate a confidence interval. This interval represents the range within which the true difference is likely to fall.

**Comparison with the Indicator Variable Model**

**Difference:** 
- The indicator variable model estimates the difference by fitting a regression model with "kind" as a categorical variable, calculating the difference in means through the modelâ€™s coefficient.

- The permutation test evaluates the significance of the observed difference by comparing it to a null distribution generated through label shuffling.

- The bootstrap confidence interval provides an interval estimate for the difference, focusing on the range rather than hypothesis testing.
