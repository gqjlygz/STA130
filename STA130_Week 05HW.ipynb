Therefore, the null hypothesis—suggesting that "nothing is happening"—appears increasingly unreasonable, or "ridiculous," because our actual data doesn't align with the expected outcomes under the null.

- Test Statistic: A number calculated from the sample data that is used to evaluate how well the data supports the null hypothesis. Examples include the t-statistic or z-statistic.
- P-value: The probability of observing a test statistic as extreme or more extreme than the one obtained from our data, assuming the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis.
- Sampling Distribution: The probability distribution of a test statistic, calculated under the assumption that the null hypothesis is true. This distribution reflects all possible values the test statistic can take, given the null hypothesis.
- Effect/Difference: A measurable change or discrepancy between groups or variables in a population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Creating the data as a CSV and reading it into a DataFrame
data = """PatientID,Age,Gender,InitialHealthScore,FinalHealthScore
1,45,M,84,86
2,34,F,78,86
3,29,M,83,80
4,52,F,81,86
5,37,M,81,84
6,41,F,80,86
7,33,M,79,86
8,48,F,85,82
9,26,M,76,83
10,39,F,83,84"""

# Writing to CSV file and reading back in (simulate the process)
with open('patient_data.csv', 'w') as file:
    file.write(data)

# Read the data
df = pd.read_csv('patient_data.csv')

# Calculate the change in health scores for each patient
df['HealthScoreChange'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Visualize the health score changes
plt.bar(df['PatientID'], df['HealthScoreChange'])
plt.axhline(0, color='grey', linestyle='--')
plt.xlabel('Patient ID')
plt.ylabel('Change in Health Score')
plt.title('Change in Health Scores After Vaccination')
plt.show()

# Bootstrapping to create a 95% confidence interval for the mean health score change
n_bootstrap = 10000
bootstrap_means = []

for _ in range(n_bootstrap):
    bootstrap_sample = np.random.choice(df['HealthScoreChange'], size=len(df['HealthScoreChange']), replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))

# Calculate the 95th percentile for a one-tailed test
upper_bound_one_tail = np.percentile(bootstrap_means, 95)

# Output result for one-sided test
print(f"95% One-Sided Bootstrap Confidence Interval: (-∞, {upper_bound_one_tail})")

# Plot the bootstrap distribution for the one-tailed test
plt.hist(bootstrap_means, bins=30, alpha=0.7, edgecolor='black')
plt.axvline(upper_bound_one_tail, color='red', linestyle='--', label='Upper 95% CI (One-Tail)')
plt.axvline(0, color='blue', linestyle='--', label='Null Hypothesis (No Effect)')
plt.xlabel('Mean Health Score Change')
plt.ylabel('Frequency')
plt.title('Bootstrap Distribution of Mean Health Score Change (One-Tailed)')
plt.legend()
plt.show()
```

### 1. **What Changed in the Code?**

For a **two-tailed test**, we test whether the observed value could be significantly greater than or less than the null hypothesis. This is reflected in the original code, which computes the **95% confidence interval** (which covers both tails, 2.5% on each side) of the bootstrap distribution.

In a **one-tailed test**, we are only interested in whether the observed value is significantly greater than (or less than) the null hypothesis in one direction. For example, if we hypothesize that the health score increased after vaccination, we would only focus on the right (positive) tail.

So, instead of calculating the 2.5th and 97.5th percentiles for a two-sided 95% confidence interval, we would calculate either the 5th percentile (for a lower-tailed test) or the 95th percentile (for an upper-tailed test).

This adjustment computes only the upper tail of the distribution, indicating how much the observed health score change exceeds what is expected under the null hypothesis.

### 2. **How This Changes the Interpretation of the Hypothesis Test:**

- In a **two-tailed test**, you are testing whether the result deviates in either direction from the null hypothesis (i.e., whether the health score could have increased or decreased significantly).
- In a **one-tailed test**, you are only interested in whether the result deviates in one direction from the null hypothesis (e.g., whether the health score **increased** after vaccination).

Thus, the one-tailed test only considers evidence in one direction. For example, in the context of this analysis, the one-tailed test would only consider the evidence that suggests the health scores improved after vaccination, not whether they got worse.

### 3. **Should We Expect the p-value to Be Smaller in a One-Tailed Test?**

Yes, we should expect the **p-value to be smaller in a one-tailed test** than in a two-tailed test, given the same data. This is because a one-tailed test concentrates all the significance in one direction, whereas a two-tailed test splits the significance between both tails of the distribution. 

In a one-tailed test, since we are only considering one tail of the distribution (either greater or less than the null hypothesis), more of the extreme values will be included in that single tail, which typically leads to a smaller p-value. However, this also means we are assuming that any deviation can only happen in one direction, so it is less conservative compared to a two-tailed test.

### Problem Introduction

**Relationship to Fisher's Original Experiment**
- Fisher’s original experiment had only 8 cups of tea, with a personalized element since Dr. Bristol claimed she had a clear preference for how tea was prepared. In contrast, the STA130 experiment uses a larger, more abstract sample of 80 students, and the task might be less personal. The nature of the population and the sample size differ, and these factors influence the statistical analysis.

**Statements of the Null Hypothesis and Alternative Hypothesis**
- Null Hypothesis (H₀): The students are randomly guessing whether the milk or tea was poured first. The probability of correctly guessing is 0.5 (p = 0.5).
- Alternative Hypothesis (H₁): More than 50% of students can correctly identify whether the tea or milk was poured first (p > 0.5).

### Quantitative Analysis

**Methodology Code and Explanations(with Visualization)**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters
n = 80  # Sample size
p_observed = 49 / 80  # Observed proportion
p_null = 0.5  # Null hypothesis proportion

# Calculate the test statistic (z-value)
z = (p_observed - p_null) / np.sqrt(p_null * (1 - p_null) / n)

# Calculate the p-value for one-tailed test
p_value = 1 - stats.norm.cdf(z)

# Print the results
print(f"Z-Statistic: {z}")
print(f"P-Value: {p_value}")

# Visualization of the Null Hypothesis Distribution
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution

# Plot the normal distribution under the null hypothesis
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Null Hypothesis (p = 0.5)', color='blue')

# Highlight the critical region (where we would reject the null hypothesis)
critical_value = stats.norm.ppf(0.95)  # 95% significance level
plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.5, label='Rejection Region (5%)')

# Add the observed test statistic line
plt.axvline(z, color='green', linestyle='--', label=f'Observed Z-Statistic: {z:.2f}')

# Add labels and legend
plt.title('Null Hypothesis Distribution and Observed Z-Statistic')
plt.xlabel('Z-Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

**Interpretation of the Visualization**

- The blue curve represents the standard normal distribution under the null hypothesis, where \( p = 0.5 \).
- The red shaded area represents the rejection region for a significance level of 5%. If the observed z-statistic lies within this area, it indicates that the observed result is significantly different from the null hypothesis.
- The green dashed line shows the observed z-statistic for this experiment. Its position relative to the red rejection region will tell us whether we can reject the null hypothesis.
